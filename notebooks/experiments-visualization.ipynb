```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import wandb
```

```python
entity = 'pintus261-university-of-trieste'
project = 'GMOOP'

api = wandb.Api()
runs = api.runs(
    f"{entity}/{project}",
    filters={
        'tags' : {'$in': ['exponential-lr']}
    }
)
```

```python
metrics = ['val_mse', 'val_mae', 'val_R2']
hyperparams = ['dropout', 'hidden_size', 'n_layers', 'activation']

data = []
for run in runs:
    # Check tags
    tags = run.tags
    if 'sinewave' in tags:
        # get the data
        row = dict()
        run_name = run.name
        generation = run_name.split('-')[1]
        individual = run_name.split('-')[2]
        row['generation'] = generation
        row['individual'] = individual
        for metric in metrics:
            row[metric] = run.summary.get(metric)
        for hyperparam in hyperparams:
            row[hyperparam] = run.config.get(hyperparam)
        data.append(row)

data = pd.DataFrame(data)
data = data.apply(pd.to_numeric)
display(data)
```
